```python
import numpy as np
import pandas as pd
from sonnia.compare_repertoires import Compare
import os
from sonnia.utils import sample_olga
```

```python
files=os.listdir('files')
files=['files/'+f for f in files]
```

```python
# generate seqs from null model
custom_pgen_model='../fig2_paper/universal_model'
gen_seqs=sample_olga(int(5e5),custom_model_folder=custom_pgen_model,chain_type='human_T_beta',vj=False)
#NB: for light chains you need to specify the parameter vj=True
```

```python
comparison=Compare(data=files,chain_type='human_T_beta',gen_seqs=gen_seqs,custom_pgen_model=custom_pgen_model)
```

```python
comparison.infer_models()
```

```python
comparison.evaluate()
```
